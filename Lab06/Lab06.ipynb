```python
# Crecimiento poblacional Ecuador y Guatemala
```

```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
Ecuador = data[data.country == "Ecuador"]
```

```python
Ecuador
```

```python
Guatemala = data[data.country == "Guatemala"]
```

```python
Guatemala
```

```python
plt.plot(Ecuador.year, Ecuador.population/10**6)
plt.plot(Guatemala.year, Guatemala.population/10**6)
plt.legend(["Ecuador", "Guatemala"])
plt.title("Poblacion Ecuador y Guatemala")
plt.xlabel("Ao")
plt.ylabel("Millones de Habitantes")
plt.show()
```

```python
Ecuador.population / Ecuador.population.iloc[0]
```

```python
Ecuador.population / Ecuador.population.iloc[0]*100
```

```python
plt.plot(Ecuador.year, Ecuador.population / Ecuador.population.iloc[0]*100)
plt.plot(Guatemala.year, Guatemala.population / Guatemala.population.iloc[0]*100)
plt.legend(["Ecuador", "Guatemala"])
plt.title("Poblacion Ecuador y Guatemala")
plt.xlabel("Ao")
plt.ylabel("Crecimiento de la poblacion (1er ao 100)")
plt.show()
```
